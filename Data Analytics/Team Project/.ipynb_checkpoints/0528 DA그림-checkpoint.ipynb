```python
#%%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

#%%
font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name, size =13 )

#%%
# 그리고 싶은 그래프에 해당하는 데이터 입력
data = pd.read_csv('C:/Users/dudwo/Desktop/19년 1학기/DA/팀과제/3.Final/상권수정_0522/의료_한의원.csv', encoding='utf8')

#%%
# 구를 기반으로 group by
gb_data = data.groupby(['gu']).count()

#%%
y_pos = np.arange(len(gb_data.index))

#%%
plt.figure(figsize=(20,10))
plt.bar(y_pos, gb_data['num'], align='center', alpha=0.5)
plt.xticks(y_pos, gb_data.index)

# 축 이름 바꾸세요
plt.ylabel('시설 수')
          
# 그래프 이름 바꾸세요
plt.title('각 구 별 한의원')
```
