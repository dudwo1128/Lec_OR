```python
#%%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

#%%
font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name, size =5 )

#%%
# 그리고 싶은 그래프에 해당하는 데이터 입력
data = pd.read_csv('C:/Users/dudwo/Desktop/연립.csv', encoding='cp949')

#%%
# 구를 기반으로 group by
gb_data = data.groupby(['건축년도']).count()

#%%
y_pos = np.arange(len(gb_data.index))

#%%
plt.figure(figsize=(20,10))
plt.bar(y_pos, gb_data['거래금액'], align='center', alpha=0.5)
plt.xticks(y_pos, gb_data.index)

# 축 이름 바꾸세요
plt.ylabel('건물 수')
          
# 그래프 이름 바꾸세요
plt.title('건축년도 별 건물 수')
```
